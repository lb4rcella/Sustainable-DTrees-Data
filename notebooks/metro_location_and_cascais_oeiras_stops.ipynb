```python
import pandas as pd
import numpy as np
import json

DATA_DIR = "../data/"
```

```python
with open(DATA_DIR+"metro_stations_lisbon.txt") as f:
  metro_json = json.load(f)
```

```python
stations_df = pd.DataFrame()
for section in metro_json["features"]:
    tmp_df = pd.DataFrame()
    tmp_df["station_name"] = [section["properties"]["INF_NOME"].rstrip("\n")]
    tmp_df["latitude"] = [section["geometry"]["coordinates"][1]]
    tmp_df["longitude"] = [section["geometry"]["coordinates"][0]]
    try:
        tmp_df["description"] = [section["properties"]["INF_DESCRICAO"].rstrip("\n")]
    except AttributeError:
        tmp_df["description"] = [section["properties"]["INF_DESCRICAO"]]
    stations_df = pd.concat([stations_df, tmp_df])
stations_df = stations_df.reset_index(drop=True)
```

```python
stations_df
```

```python
stations_df.to_csv(DATA_DIR+"metro_location.csv", index=False)
```

## Process Cascais bus stops txt

```python
cascais_stops_raw = pd.read_csv(DATA_DIR+"cascais_bus_stops_raw.txt")
```

```python
cascais_stops_raw = cascais_stops_raw.rename(columns={"stop_lat": "latitude", "stop_lon": "longitude"})
cascais_stops_raw = cascais_stops_raw.loc[:, ["stop_id", "stop_name", "latitude", "longitude", "wheelchair_boarding"]]
cascais_stops_raw["wheelchair_boarding"] = np.where(cascais_stops_raw["wheelchair_boarding"] == 1,True,False)
```

```python
cascais_stops_raw
```

```python
cascais_stops_raw.to_csv(DATA_DIR+"cascais_bus_stops.csv", index=False)
```

```python
oeiras_stops_converted = pd.read_csv(DATA_DIR+"oeiras_bus_stops_raw_converted.csv")
```

```python
oeiras_stops_converted = oeiras_stops_converted.rename(columns={"X": "longitude", "Y": "latitude", 'gid': 'stop_id', 
                                                                'Estado de Conservação': 'Conditions'})
oeiras_stops_converted = oeiras_stops_converted.loc[:, ["stop_id", "latitude", "longitude", "Conditions"]]
oeiras_stops_converted.loc[oeiras_stops_converted.Conditions == "Bom", "Conditions"] = "Good"
oeiras_stops_converted.loc[oeiras_stops_converted.Conditions == "Razoável", "Conditions"] = "Average"
oeiras_stops_converted.loc[oeiras_stops_converted.Conditions == "Mau", "Conditions"] = "Bad"
```

```python
oeiras_stops_converted.to_csv(DATA_DIR+"oeiras_bus_stops.csv", index=False)
```
