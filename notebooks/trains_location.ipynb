```python
from bs4 import BeautifulSoup
import requests
import unidecode
import pandas as pd
import numpy as np
```

```python
url="https://www.cp.pt/passageiros/pt/consultar-horarios/estacoes"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
```

```python
count = 0
stations = list()
for link in soup.find_all("a"):
    if 45 <= count <= 406:
        stations.append(unidecode.unidecode(link.text.replace(" ", "-").lower()))
    count += 1
```

```python
stations
```

```python
new_url = f'https://www.cp.pt/passageiros/pt/consultar-horarios/estacoes/{stations[0].lower()}'
```

```python
coords_per_station = dict()


for each_station in stations:
    
    new_url = f'https://www.cp.pt/passageiros/pt/consultar-horarios/estacoes/{each_station}'
    
    html_content = requests.get(new_url).text
    print(new_url)

    # Parse the html content
    soup = BeautifulSoup(html_content, "lxml")
    count = False

    for i in soup.prettify().split():
        if 'Coordenadas:' in i:
            count = 0

        if count >= 0:
            count += 1

        if count ==3:
            coords_per_station[each_station] = i.split("|")
            continue
```
